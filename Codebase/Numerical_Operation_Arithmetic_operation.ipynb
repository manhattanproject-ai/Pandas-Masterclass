# 1. Numerical data - Arithmetic operation

* Arithmetic Operation: Covers standard operations like Addition, Subtraction, Multiplication (including Basic and Dot Product), Division (Floating point and Integer point), Modulo, and Power. These operations also allow for filling missing values, selecting the axis of operation, and working with multi-index levels.
* Reverse Arithmetic Operation: Lists the reverse counterparts of these standard operations, specifically reverse add, reverse subtraction, reverse multiplication, reverse division, reverse modulo, and reverse power.

### 2. Import the required dataset and perform initial analysis

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('Movies.csv')
movies.head()
```

```python
movies.shape
```

```python
# View numerical columns of movies dataset

df1 = movies.select_dtypes(include='number')
df1.head()
```

#3. Arithmetic operation - addition

* adding a number to a specific column
* adding a number to all column
* filling NaN values before operation
* Adding different numbers to different columns
* Apply addition of list of scalars at index axis

### A. adding a number to a specific column

```python
# Lets visualize the numerical column (num_critic_for_reviews column) of interest

df1['num_critic_for_reviews'].head()
```

```python
# Add 10 to num_critic_for_reviews column - Addition operation

df1['num_critic_for_reviews'] = df1['num_critic_for_reviews'].add(10)
df1[['num_critic_for_reviews']].head()
```

### B. adding a number to all column

```python
# Add 10 to all columns

df1 = df1.add(10)
df1.head()
```

### C. filling NaN values before operation

```python
round((df1.isnull().sum()/len(df1.index))*100,2)
```

```python
# As you can see 1 NaN is still their in num_critic_for_review column , since additional operation can't be performed on NaN values
```

```python
# Specifying fill value before addition
# first the nan's are converted to 0 and then 10 is added , if you directly want to add a scaler to a nan value it will not work

df1 = df1.add(10, fill_value=0)

df1.head()
```

```python
round((df1.isnull().sum()/len(df1.index))*100,2)
```

### D. Adding different numbers to different columns

```python
df2 = movies.select_dtypes(include='number')
df2.head()
```

```python
# Creating a small dataframe with 1st 4 columns
df3 = df2.iloc[: , :4]
df3.head()
```

```python
# Apply addition (adding different values to different columns - 10 , 30 , 50 , 70 respectively )

df_added = df3.add([10, 30, 50, 70], axis='columns')

df_added.head()
```

### E. Apply addition of list of scalars at index axis

```python
df_added.head()
```

```python
# Filter out top 5 rows from df_added and create a new dataframe called df_r

df_r = df_added.iloc[:5 , :]
df_r
```

```python
# Apply addition of list of scalars at index axis

df_added1 = df_r.add([10, 30, 50, 70, 90], axis='index')
df_added1
```

# 6. Arithmetic operation - division

* Floating point division
* Integer point division

### A. Floating point division

```python
# View numerical columns of movies dataset

df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Apply floating point division

df_div = df2.div(4)
df_div.head()
```

### B. Integer point division

```python
# View numerical columns of movies dataset

df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Apply integer division
df_div = df2.floordiv(4)

print('After integer division')
df_div.head()
```

# 7. Arithmetic operation - Modulo

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Create pandas Series object used for division

series_other = pd.Series([1, 2, 3, 4 , 5],
                          index=['num_critic_for_reviews', 'duration',
                                'director_facebook_likes', 'actor_3_facebook_likes' , 'actor_1_facebook_likes'])

# Apply modulo operation
df_mod = df2.mod(series_other)

df_mod.head()
```

# 8. Arithmetic operation - Power

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Dictionary with column names and value to raise by

power_dict = {'num_critic_for_reviews':2,
              'duration':3,
              'director_facebook_likes':2,
              'actor_3_facebook_likes':3,
              'actor_1_facebook_likes':2
            }

# Calculate exponential power
df_power = df2.pow(power_dict)

df_power.head()
```
