# 3. Largest N

* df.nlargest(n, columns='column_name') : Selects the n rows with the largest values in the specified single column.
* df.nlargest(n, columns=['col1', 'col2']) : Selects the n rows with the largest values, prioritizing col1 first, then col2 for ties, and so on for multiple specified columns.
* df.nlargest(n, columns='column_name', keep='first') : Selects the n rows with the largest values, keeping the first occurring rows if there are ties.
* df.nlargest(n, columns='column_name', keep='last') : Selects the n rows with the largest values, keeping the last occurring rows if there are ties.
* df.nlargest(n, columns='column_name', keep='all') : Selects the n rows with the largest values, keeping all rows if there are ties that would result in more than n rows.

### A. df.nlargest(n, columns='column_name') : Selects the n rows with the largest values in the specified single column.

```python
# Obtain top 5 directors by num_critic_for_reviews

df_largest = df1.nlargest(n=5, columns='num_critic_for_reviews')

df_largest
```

### B. df.nlargest(n, columns=['col1', 'col2']) : Selects the n rows with the largest values, prioritizing col1 first, then col2 for ties, and so on for multiple specified columns.

```python
# Obtain top 6 directors by num_critic_for_reviews and imdb_score
# The ordering will ensure to rank by "num_critic_for_reviews" , when its a tie they will use "imdb_score"

df_largest = df1.nlargest(6, ['num_critic_for_reviews', 'imdb_score'])

df_largest
```

### C. df.nlargest(n, columns='column_name', keep='first') : Selects the n rows with the largest values, keeping the first occurring rows if there are ties.

```python
# Obtain top 4 directors by num_critic_for_reviews and duration
# Application on duplicate rows using keep

# Keep 'first' occurence
df_first = df1.nlargest(4, ['num_critic_for_reviews', 'duration'] , keep='first')

df_first
```

### D. df.nlargest(n, columns='column_name', keep='last') : Selects the n rows with the largest values, keeping the last occurring rows if there are ties.

```python
# Obtain top 4 directors by num_critic_for_reviews and duration
# Application on duplicate rows using keep

# Keep 'last' occurence

df_last = df1.nlargest(4, ['num_critic_for_reviews', 'duration'] , keep='last')

df_last
```

### E. df.nlargest(n, columns='column_name', keep='all') : Selects the n rows with the largest values, keeping all rows if there are ties that would result in more than n rows.

```python
# Obtain top 4 directors by num_critic_for_reviews and duration
# Application on duplicate rows using keep

# Keep 'all' occurence
df_all = df1.nlargest(4, ['num_critic_for_reviews', 'duration'] , keep='all')

df_all
```

# 5. Rank

The rank() method in Pandas assigns a numerical rank to each value within a Series or DataFrame based on its position in the sorted order of values. It's particularly useful for understanding the relative standing of each data point, such as identifying the top-performing items or segmenting data into quartiles. The method offers various strategies to handle tied values (e.g., assigning the average rank, minimum rank, or sequential rank based on appearance) and options for handling missing data, providing flexibility for different analytical needs.

* df['column_name'].rank() : Assigns a rank to each value in the Series, with ties receiving the average rank (default 'average' method).
* df['column_name'].rank(method='average') : Assigns average rank to values that are equal (ties).
* df['column_name'].rank(method='min') : Assigns the minimum rank to values that are equal (ties).
* df['column_name'].rank(method='max') : Assigns the maximum rank to values that are equal (ties).
* df['column_name'].rank(method='first') : Assigns ranks in the order of appearance for values that are equal (ties).
* df['column_name'].rank(method='dense') : Assigns ranks without gaps, assigning the same rank to equal values.
* df['column_name'].rank(ascending=False) : Assigns ranks in descending order, where the largest value gets rank 1.

```python
df2 = df1.tail(20)
```

```python
# Sort by rating column
df2 = df2.sort_values(by='duration')

# Rank method - Average
df2['duration_rank_avg'] = df2['duration'].rank(method='average')

# Rank method - Minimum
df2['duration_rank_min'] = df2['duration'].rank(method='min')

# Rank method - Maximum
df2['duration_rank_max'] = df2['duration'].rank(method='max')

# Rank method - First
df2['duration_rank_first'] = df2['duration'].rank(method='first')

# Rank method - Dense
df2['duration_rank_dense'] = df2['duration'].rank(method='dense')

# Filter to relevant columns
df2 = df2[['duration', 'duration_rank_avg', 'duration_rank_min',
         'duration_rank_max', 'duration_rank_first', 'duration_rank_dense' ]]

# Display all rows
df2
```
