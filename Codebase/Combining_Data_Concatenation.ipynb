# 1. Data Concatenation

Imagine the table in the image is a collection of data, like a spreadsheet.

* Rows: Represent individual records or entries (like information about different people).
* Columns: Represent different attributes or features (like ID, Gender, DOB, City, Postal). What is Concatenation?
* Concatenation in Pandas is like "gluing" or "stacking" two or more DataFrames (which are essentially tables) together. You can do this in two main ways:

Concatenating Along Rows (Stacking Vertically):

* This is like adding more rows to the bottom of your existing table.
* You're essentially appending new data records.
* In Pandas, this is done along axis=0.

Concatenating Along Columns (Stacking Horizontally):

* This is like adding more columns to the right of your existing table.
* You're essentially adding more attributes or features.
* In Pandas, this is done along axis=1.

Key Points

* axis Parameter: The axis parameter determines whether to concatenate along rows (0) or columns (1).
* Index Alignment: Pandas will align the DataFrames based on their indexes (row labels) or column names, depending on the axis.
* Missing Values: If the DataFrames have different columns or indexes, Pandas may introduce missing values (NaN) to fill in the gaps.

Summary

Concatenation is like combining multiple spreadsheets into one. You can either stack them on top of each other (adding more rows) or place them side by side (adding more columns). Pandas provides the pd.concat() function to do this easily.

# 2. Concatenating along Rows

Concatenating across rows is like stacking two spreadsheets on top of each other. You're adding more rows to your data while keeping the same columns.

key criteria for Concatenation across rows in Pandas:

* Shared Columns: DataFrames should ideally have the same or highly similar column names for a clean merge.
* Vertical Stacking: The goal is to append rows from one DataFrame to another.
axis=0: This is the default axis for pd.concat(), indicating row-wise concatenation.
* Index Handling: Consider ignore_index=True if original indices are not meaningful or would create duplicates.
* NaN Introduction: Non-matching columns between DataFrames will result in NaN values in the output.
* Data Types: Columns with the same name across DataFrames should ideally have compatible data types to avoid unexpected type coercion.

### A. Import necessary Dataset

```python
import pandas as pd

april = pd.read_excel('Book_Sales_April.xlsx')
may = pd.read_excel('Book_Sales_May.xlsx')
june = pd.read_excel('Book_Sales_June.xlsx')
```

```python
print(april.shape)

print('='*100)

april.head()
```

```python
print(may.shape)

print('='*100)

may.head()
```

```python
print(june.shape)

print('='*100)

june.head()
```

### B. Row wise concatenation of april and may dataframe

```python
# rowwise concatenation of two dataframes

df_apr_may = pd.concat([april , may] , axis=0 , ignore_index=True)
```

```python
# visualize the shape and structure of concatenated data
print(df_apr_may.shape)

print('='*100)

df_apr_may.head()
#df_apr_may.tail()
```

### C. Row wise concatenation of 3 dataframes

```python
# rowwise concatenation of 3 dataframes

df_apr_may_june = pd.concat([april , may , june] , axis=0 , ignore_index=True)
```

```python
print(df_apr_may_june.shape)

print('='*100)

df_apr_may_june.head()
#df_apr_may_june.tail()
```

# 3. Concatenation across columns

Concatenating across columns is like merging two spreadsheets side by side. You're adding more columns to your data while keeping the same rows.

key criteria for Concatenation across columns in Pandas:

* Shared Indices: DataFrames should ideally have the same or highly similar index labels for a clean merge.
* Horizontal Stacking: The goal is to append columns from one DataFrame next to columns of another.
* axis=1: This parameter is required for pd.concat() to indicate column-wise concatenation.
* Index Alignment: Pandas aligns rows based on their index labels.
* NaN Introduction: Non-matching index labels between DataFrames will result in NaN values in the output.
* Column Uniqueness: Column names across DataFrames should ideally be distinct if you want all original columns to appear without ambiguity.
