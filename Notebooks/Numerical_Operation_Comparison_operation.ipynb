# 3. Different types of Comparison operation - baseline example

### A. Comparison operation - Equal to (==)

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Define scalar for comparison
scalar = 1000

# Show result of comparison
output = df2.eq(scalar)

output.head()
```

```python
# looks for 1000 as a value if the value matches 1000 makes it True , else makes it False
```

### B. Comparison operation - Not Equal to (!=)

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Define scalar for comparison
scalar = 1000

# Show result of comparison
output = df2.ne(scalar)

output.head()
```

```python
# looks for 1000 as a value if the value matches 1000 makes it False , else makes it True
```

### C. Comparison operation - Less than to equal to (<=**bold text**)

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Define scalar for comparison
scalar = 1000

# Show result of comparison
output = df2.le(scalar)

output.head()
```

```python
# looks for 1000 as a value if the value is less than or equal to  1000 makes it True , else makes it False
```

### D. Comparison operation - Less than (<)

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Define scalar for comparison
scalar = 1000

# Show result of comparison
output = df2.lt(scalar)

output.head()
```

```python
# looks for 1000 as a value if the value is less than  1000 makes it True , else makes it False
```

### E. Comparison operation - Greater than or equal to (>=)

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Define scalar for comparison
scalar = 1000

# Show result of comparison
output = df2.ge(scalar)

output.head()
```

```python
# looks for 1000 as a value if the value is greater than or equal to  1000 makes it True , else makes it False
```

### F. Comparison operation - Greater than (>)

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

# 4. Different types of Comparison operation - advanced example

### A. Compare different columns with different scaler values

```python
df1 = movies.select_dtypes(include='number')
df2 = df1.iloc[: , :5]
df2.head()
```

```python
# Define list for comparison
# Since axis of comparison is columns , the list size should match the number of columns

lst = [723, 178 , 563 , 530 , 640]

# Show result of comparison
output = df2.eq(lst, axis='columns')

output.head()
# Column 1 compared with 723
# Column 2 compared with 178
# Column 3 compared with 563
# Column 4 compared with 530
# Column 5 compared with 640
# In case value matches its True , else False
```

```python
# Define list for comparison
# Since axis of comparison is columns , the list size should match the number of columns

lst = [723, 178 , 563 , 530 , 640]

# Show result of comparison
output = df2.ne(lst, axis='columns')

output.head()
# Column 1 compared with 723
# Column 2 compared with 178
# Column 3 compared with 563
# Column 4 compared with 530
# Column 5 compared with 640
# In case value matches its False , else True
```

```python
# Define list for comparison
# Since axis of comparison is columns , the list size should match the number of columns

lst = [723, 178 , 563 , 530 , 640]

# Show result of comparison
output = df2.le(lst, axis='columns')

output.head()
# Column 1 compared with 723
# Column 2 compared with 178
# Column 3 compared with 563
# Column 4 compared with 530
# Column 5 compared with 640
# In case value matches or lesses its True , else False
```

```python
# Define list for comparison
# Since axis of comparison is columns , the list size should match the number of columns

lst = [723, 178 , 563 , 530 , 640]

# Show result of comparison
output = df2.lt(lst, axis='columns')

output.head()
# Column 1 compared with 723
# Column 2 compared with 178
# Column 3 compared with 563
# Column 4 compared with 530
# Column 5 compared with 640
# In case value is lesses its True , else False
```

```python
# Define list for comparison
# Since axis of comparison is columns , the list size should match the number of columns

lst = [723, 178 , 563 , 530 , 640]

# Show result of comparison
output = df2.ge(lst, axis='columns')

output.head()
# Column 1 compared with 723
# Column 2 compared with 178
# Column 3 compared with 563
# Column 4 compared with 530
# Column 5 compared with 640
# In case value is greater or equal its True , else False
```

```python
# Define list for comparison
# Since axis of comparison is columns , the list size should match the number of columns

lst = [723, 178 , 563 , 530 , 640]

# Show result of comparison
output = df2.gt(lst, axis='columns')

output.head()
# Column 1 compared with 723
# Column 2 compared with 178
# Column 3 compared with 563
# Column 4 compared with 530
# Column 5 compared with 640
# In case value is greater its True , else False
```
