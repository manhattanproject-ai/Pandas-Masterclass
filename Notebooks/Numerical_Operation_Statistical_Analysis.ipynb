# 3. Statistical Analysis - Central tendency of the data

The central tendency of a dataset refers to a single, representative value that aims to describe the "center" or "typical" position of the data within its distribution. It summarizes the most characteristic or common score in a dataset, providing a concise measure around which the data points tend to cluster. The most common measures of central tendency are the mean (average), median (middle value), and mode (most frequent value), each offering a different perspective on the dataset's typical value.

### C. Median of the columns

The median is the middle value in a dataset that has been sorted in ascending order. It separates the lower half of the data from the upper half.

Here's how to calculate the median:

* Sort the data: Arrange the values in ascending order (from smallest to largest).
* Find the middle:
  * If the dataset has an odd number of values: The median is the middle value. * If the dataset has an even number of values: The median is the average of the two middle values.
  
Examples:

* Odd number of values:

Dataset: 22, 25, 28, 30, 35

Sorted data: 22, 25, 28, 30, 35

Median: 28

* Even number of values:

Dataset: 22, 25, 28, 30

Sorted data: 22, 25, 28, 30 Median: (25 + 28) / 2 = 26.5

Important Notes:

* The median is less affected by outliers than the mean.
* Extreme values don't influence the middle value as much.
* The median is often used for datasets with skewed distributions, where the mean might be misleading.
* In Pandas, you can calculate the median of a DataFrame or Series using the df.median() method.
*  It handles NaN (Not a Number) values by ignoring them.

image.png

```python
print('Median value of Age:' , df['Age'].median())
print('Median value of fnlwgt:' , df['fnlwgt'].median())
print('Median value of Education-Num:' , df['Education-Num'].median())
print('Median value of Capital Gain:' , df['Capital Gain'].median())
print('Median value of Capital Loss:' , df['Capital Loss'].median())
print('Median value of Hours per week:' , df['Hours per week'].median())
```

### D. Mode

The mode is the value that appears most frequently in a dataset. A dataset can have one mode (unimodal), more than one mode (multimodal), or no mode at all if all values occur with the same frequency.

Here's how to calculate the mode:

Count occurrences: Determine how many times each value appears in the dataset. Identify the most frequent: The value(s) with the highest count is/are the mode(s).

Examples:

* Unimodal:

Dataset: 22, 25, 28, 30, 35, 25

* 22 appears once.
* 25 appears twice.
* 28 appears once.
* 30 appears once.
* 35 appears once.

Mode: 25

* Multimodal:

Dataset: 22, 25, 28, 30, 25, 28

* 22 appears once.
* 25 appears twice.
* 28 appears twice.
* 30 appears once.

Modes: 25 and 28

* No mode:

Dataset: 22, 25, 28, 30

All values appear once.

Mode: None

Important Notes:

The mode is the only measure of central tendency that can be used for categorical data (e.g., colors, types of fruit). A dataset can have no mode, one mode, or multiple modes. In Pandas, you can calculate the mode of a DataFrame or Series using the df.mode() method.

# 4. Statistical Analysis - Measures of dispersion or variability

Measures of dispersion (or variability) quantify the spread or scattering of data points around the central tendency of a dataset. Instead of describing where the data is centered, these metrics indicate how much the individual data values differ from each other and from the average. A high measure of dispersion suggests that data points are widely spread out, implying greater diversity or volatility, while a low measure indicates that data points are tightly clustered together, suggesting consistency or homogeneity.

### B. Standard Deviation

Example:

Let's say you have the following dataset: 2, 4, 6, 8

Calculate the Mean: (2 + 4 + 6 + 8) / 4 = 5

Calculate the Variance:

* (2 - 5)² = 9
* (4 - 5)² = 1
* (6 - 5)² = 1
* (8 - 5)² = 9
* (9 + 1 + 1 + 9) / 4 = 5 (Variance)

Calculate the Standard Deviation:

√5 ≈ 2.236

Key points:

A higher standard deviation means the data is more spread out. A lower standard deviation means the data is more clustered around the mean. Standard deviation is very important in statistical analysis, helping us understand the distribution of data.

### C. Variation

Example:

Let's say your dataset is: [2, 4, 6, 8]

Calculate the Mean: (2 + 4 + 6 + 8) / 4 = 5

Calculate the Squared Differences:

* (2 - 5)² = 9
* (4 - 5)² = 1
* (6 - 5)² = 1
* (8 - 5)² = 9

Calculate the Variance:

(9 + 1 + 1 + 9) / 4 = 5

So, the variance of the dataset [2, 4, 6, 8] is 5.

Key Points:

Variance measures how much the data points deviate from the mean. A higher variance indicates that the data is more spread out. A lower variance indicates that the data is more clustered around the mean. Variance is always a non-negative value. The square root of the variance gives you the standard deviation, which is another common measure of dispersion.

image.png

```python
print('Variation of Age:' , df['Age'].var())
print('Variation of fnlwgt:' , df['fnlwgt'].var())
print('Variation of Education-Num:' , df['Education-Num'].var())
print('Variation of Capital Gain:' , df['Capital Gain'].var())
print('Variation of Capital Loss:' , df['Capital Loss'].var())
print('Variation of Hours per week:' , df['Hours per week'].var())
```

### Interpretation

* Variation of Age (186.06): This indicates a moderate spread in the ages of individuals. Since the unit of variance is squared, it's not directly interpretable in "years," but relative to other variables, it suggests a reasonable diversity in age within the dataset.
* Variation of fnlwgt (11,140,797,791.84): This is an extremely large variance. Given fnlwgt often represents population estimates, this huge number signifies an enormous spread in the "final weight" values, indicating a very diverse representation of individuals per record.
* Variation of Education-Num (6.62): This is a very small variance, which suggests that the number of years of education for individuals in the dataset is quite clustered around the mean (around 10.1 years), indicating relatively little variability in educational attainment.
* Variation of Capital Gain (54,542,539.18): This is a very large variance. Given that the median capital gain was 0, this large variance strongly implies a highly skewed distribution with a few individuals having extremely large capital gains that pull the variance up significantly.
* Variation of Capital Loss (162,376.94): While much smaller than capital gain variance, this is still a substantial number, indicating a considerable spread in capital loss values among those who experience losses, even though most have zero loss.
* Variation of Hours per week (152.46): This represents a moderate level of spread in the hours worked per week. It suggests that while many work around 40 hours, there's a notable dispersion, with some working significantly more or less.

General Note on Variance: Variance is a squared unit, which can make direct interpretation less intuitive than its square root, the standard deviation. A larger variance always means greater dispersion of data points around the mean.

### Standard Deviation vs Variance

When to use what:

1. Standard Deviation:

* When you need to understand the spread of data in the context of the original units.
* When communicating variability to a non-technical audience.
* When comparing the variability of different datasets with the same units.

2. Variance:

* When performing statistical calculations or modeling, as it has convenient mathematical properties.
* When comparing the variability of different datasets with different units (since the units are squared, they become irrelevant for comparison).

In essence:

Standard deviation is more interpretable. Variance is more mathematically useful.

Example:

If you're analyzing the heights of students, standard deviation (e.g., "The standard deviation is 5 cm") would be easier to understand than variance (e.g., "The variance is 25 cm²"). However, variance might be preferred in certain statistical formulas.

Ultimately, the choice depends on the specific context and purpose of your analysis.

### D. Quartile Analysis

* Dataset: 6, 7, 15, 36, 39, 40, 41, 42, 43, 47, 49

Key Points:

Quartiles help understand the spread and distribution of data. The Interquartile Range (IQR) is Q3 - Q1 and represents the spread of the middle 50% of the data. Quartiles are used for box plots and identifying potential outliers.

image.png


Q1, Q2, Q3, and IQR are all measures of position that divide a dataset into groups, providing insights into the distribution and spread of the data. Here's the significance of each:

1. Q1 (First Quartile):
* Marks the 25th percentile of the data.
* 25% of the values in the dataset fall below Q1.
* Helps identify the lower end of the data's spread.

2. Q2 (Second Quartile/Median):
* Marks the 50th percentile of the data.
* Divides the data in half, with 50% of values falling below and 50% above.
* Represents the middle value of the dataset.

3. Q3 (Third Quartile):
* Marks the 75th percentile of the data.
* 75% of the values in the dataset fall below Q3.
* Helps identify the upper end of the data's spread.

4. IQR (Interquartile Range):
* Calculated as Q3 - Q1.
* Represents the range of the middle 50% of the data.
* Less affected by extreme values (outliers) than the overall range.

Used to:

* Identify potential outliers.
* Understand the variability of the data.
* Construct box plots for visualizing the distribution.

Overall Significance:

* Provide a more detailed picture of the data's distribution than just the mean or median.
* Help identify where the majority of the data lies.
* Useful for comparing distributions across different datasets.
* Essential for understanding the spread and variability of data.

### E. Standard Error of the Mean (SEM)

The Standard Error of the Mean (SEM) is a statistical measure that quantifies the accuracy with which a sample mean represents the true population mean.

In simpler terms, if you were to take many different samples from the same population and calculate the mean for each sample, the SEM tells you how much those sample means are likely to vary from the actual mean of the entire population. A smaller SEM indicates that the sample mean is a more precise estimate of the population mean.

image.png

Key points about SEM:

Measure of Precision: It's a measure of the precision of the sample mean, not the variability of the individual data points themselves (that's what standard deviation measures).
Decreases with Sample Size: As the sample size (n) increases, the SEM decreases. This makes intuitive sense: larger samples generally provide a more reliable estimate of the population mean.
Used for Confidence Intervals: SEM is often used to calculate confidence intervals for the population mean, which provide a range within which the true population mean is likely to fall.

```python
print('SEM of Age:' , round(df['Age'].sem(),2))
print('SEM of fnlwgt:' , round(df['fnlwgt'].sem(),2))
print('SEM of Education-Num:' , round(df['Education-Num'].sem(),2))
print('SEM of Capital Gain:' , round(df['Capital Gain'].sem(),2))
print('SEM of Capital Loss:' , round(df['Capital Loss'].sem(),2))
print('SEM of Hours per week:' , round(df['Hours per week'].sem(),2))
```

# 5. Statistical Analysis - Understanding of Distribution

A distribution in statistics describes the arrangement of data points, showing how frequently each value or range of values appears within a dataset. Understanding a distribution is crucial because it provides fundamental insights into the nature of the data: it reveals where data is concentrated (central tendency), how spread out it is (variability), whether it's symmetrical or skewed, and if outliers exist. This knowledge is essential for choosing appropriate statistical methods, making informed interpretations, identifying patterns or anomalies, and building more accurate models, ultimately guiding better decision-making from the data.

### A. Skewness -

Skewness is a measure of the asymmetry of the distribution of a dataset. Here's how it's calculated:

Calculate the Mean (μ):

Find the average of all the values in the dataset.

Calculate the Standard Deviation (σ):

A measure of how spread out the data is around the mean (see previous explanations for how to calculate this).

Calculate the Skewness:
This involves calculating the third standardized moment. Here's the formula:

Skewness = Σ(xi - μ)³ / (N * σ³)

Where:

* xi = each value in the dataset
* μ = the mean of the dataset
* σ = the standard deviation of the dataset
* N = the number of values in the dataset

Interpretation:

* Skewness = 0: The distribution is perfectly symmetrical.
* Skewness > 0 (Positive Skew): The distribution has a longer tail on the right side. The mean is greater than the median.
* Skewness < 0 (Negative Skew): The distribution has a longer tail on the left side. The mean is less than the median.

In simpler terms:

Imagine a bell curve. If the bell curve is stretched towards the right, it's positively skewed. If it's stretched towards the left, it's negatively skewed.

Why is skewness important?

* It helps understand the shape of the data distribution.
* Many statistical models assume a normal distribution (symmetrical). Skewness helps assess if this assumption is valid.
* It can influence the choice of appropriate summary statistics (mean vs. median).

```python
print('Skewness of Age:' , df['Age'].skew())
print('Skewness of fnlwgt:' , df['fnlwgt'].skew())
print('Skewness of Education-Num:' , df['Education-Num'].skew())
print('Skewness of Capital Gain:' , df['Capital Gain'].skew())
print('Skewness of Capital Loss:' , df['Capital Loss'].skew())
print('Skewness of Hours per week:' , df['Hours per week'].skew())
```

### B. Kurtosis

Kurtosis is a measure of the "tailedness" and peakedness of the distribution of a dataset. Here's how it's calculated:

* Calculate the Mean (μ):
* Find the average of all the values in the dataset.

Calculate the Standard Deviation (σ):

* A measure of how spread out the data is around the mean.

Calculate the Kurtosis:

This involves calculating the fourth standardized moment. Here's the formula:

Kurtosis = Σ(xi - μ)⁴ / (N * σ⁴)

Where:

* xi = each value in the dataset
* μ = the mean of the dataset
* σ = the standard deviation of the dataset
* N = the number of values in the dataset

Interpretation:

* Kurtosis = 3 (or 0 in some normalizations): The distribution is mesokurtic, similar to a normal distribution.
* Kurtosis > 3 (Leptokurtic): The distribution has heavier tails and a sharper peak than a normal distribution. The data is more concentrated around the mean, with more outliers.
* Kurtosis < 3 (Platykurtic): The distribution has lighter tails and a flatter peak than a normal distribution. The data is less concentrated around the mean.

In simpler terms:

* Imagine a bell curve.
* A leptokurtic distribution is like a tall, skinny bell with long tails.
* A platykurtic distribution is like a wide, flat bell with short tails.

Why is kurtosis important?

It helps understand the shape of the data distribution, particularly how extreme values are distributed. It can indicate the presence of outliers. It's used in various statistical tests and models.

```python
print('Kurtosis of Age:' , df['Age'].kurt())
print('Kurtosis of fnlwgt:' , df['fnlwgt'].kurt())
print('Kurtosis of Education-Num:' , df['Education-Num'].kurt())
print('Kurtosis of Capital Gain:' , df['Capital Gain'].kurt())
print('Kurtosis of Capital Loss:' , df['Capital Loss'].kurt())
print('Kurtosis of Hours per week:' , df['Hours per week'].kurt())
```

* Kurtosis of fnlwgt: 6.219
  * This value is significantly greater than 0. It indicates a leptokurtic distribution, meaning it has heavier tails and a sharper, more peaked center than a normal distribution. This is consistent with its high positive skewness; the large positive values in the tails contribute to both the skew and the high kurtosis.
* Kurtosis of Education-Num: 0.623

It suggests that the Education-Num distribution is mesokurtic to slightly leptokurtic, meaning its tails are slightly heavier and its peak is slightly sharper than a normal distribution. In summary: High positive kurtosis values (like fnlwgt, Capital Gain, Capital Loss) indicate that extreme values (outliers) are more common than in a normal distribution, contributing significantly to the overall spread, often seen in highly skewed data. # 6. Statistical Analysis - Pairwise relationship

A pairwise relationship in a dataset refers to the association or connection observed between any two individual variables within that dataset, examined independently of all other variables. It describes how changes in one variable correspond to changes in another, indicating patterns such as positive or negative correlation, or the absence of any linear or monotonic trend, when considering only those two variables at a time.

### A. Covariance

Covariance is a statistical measure that describes the directional relationship between two random variables. It tells us whether two variables tend to move in the same direction, in opposite directions, or if there's no consistent linear relationship between them.

* Positive Covariance: If the covariance is positive, it means that as one variable increases, the other variable also tends to increase. Similarly, if one decreases, the other tends to decrease. They move in the same direction.
* Negative Covariance: If the covariance is negative, it means that as one variable increases, the other variable tends to decrease, and vice-versa. They move in opposite directions.
* Near-Zero Covariance: If the covariance is close to zero, it suggests that there is no strong linear relationship between the two variables. They might move independently, or there might be a non-linear relationship that covariance doesn't capture.

Example:

Let's consider a simple dataset for a few students, looking at two variables: Daily Hours Spent Studying and their Exam Score.

They show a consistent pattern of moving in the same direction.

If, however, we were comparing "Hours Spent Watching TV" and "Exam Score," we might find a negative covariance, implying that as TV watching increases, exam scores tend to decrease.

### B. Correlation

Correlation is a statistical measure that quantifies the strength and direction of a linear relationship between two numerical variables. Unlike covariance, correlation is standardized, meaning its value always falls within a specific range, making it much easier to interpret.

The correlation coefficient (often denoted as 'r' or 'ρ') ranges from -1 to +1:

* +1 (Perfect Positive Correlation): As one variable increases, the other variable increases proportionally in the same direction. All data points would fall perfectly on an upward-sloping straight line.
* -1 (Perfect Negative Correlation): As one variable increases, the other variable decreases proportionally. All data points would fall perfectly on a downward-sloping straight line.
* 0 (No Linear Correlation): There is no consistent linear relationship between the two variables. Changes in one variable do not predict consistent changes in the other. (Important: This doesn't mean there's no relationship at all, just no linear one; there could be a non-linear pattern).
* Values between 0 and +1 (Positive Correlation): A positive relationship exists, but it's not perfect. Higher values closer to +1 indicate a stronger positive linear relationship.
* Values between -1 and 0 (Negative Correlation): A negative relationship exists, but it's not perfect. Lower values closer to -1 indicate a stronger negative linear relationship.

DIFFERENT TYPES OF CORRELATION -

1. Pearson Correlation Coefficient (Pearson's r)

* What it Measures: Pearson correlation measures the strength and direction of a linear relationship between two continuous numerical variables. It assesses how well a straight line can describe the relationship between two variables.

Key Characteristics/Assumptions:

* Linearity: Assumes the relationship between the variables is linear.
Continuous Data: Both variables should be measured on an interval or ratio scale (e.g., temperature, height, income).
* Normality (for inference): For hypothesis testing and confidence intervals, it assumes that the data for each variable is approximately normally distributed.
* Homoscedasticity: Assumes the variance of one variable is roughly equal across all levels of the other variable.
* Sensitivity to Outliers: Pearson correlation can be heavily influenced by outliers, which can distort the true relationship.
* When to Use It: When you want to find out if there's a straight-line relationship between two numerical variables that are approximately normally distributed, and you are confident there are no extreme outliers pulling the relationship.
* Example: The linear relationship between a person's height and their weight, or between advertising spending and sales revenue.

2. Spearman's Rank Correlation Coefficient (Spearman's ρ )

* What it Measures: Spearman correlation measures the strength and direction of a monotonic relationship between two variables. Instead of using the raw data values, it converts the data into ranks and then calculates the Pearson correlation on these ranks. A monotonic relationship means that as one variable increases, the other variable either consistently increases (but not necessarily linearly) or consistently decreases.

Key Characteristics/Assumptions:

* Monotonicity: The relationship must be monotonic (either consistently increasing or consistently decreasing), but not necessarily linear.
* Ordinal Data: Suitable for ordinal data (ranked data) or when interval/ratio data doesn't meet Pearson's assumptions (e.g., non-normal distribution, presence of outliers).
* Non-Parametric: It is a non-parametric test, meaning it doesn't assume any specific distribution for the data.
* Less Sensitive to Outliers: Because it uses ranks, it is less affected by extreme outliers compared to Pearson's correlation.
* When to Use It: When you have ordinal data, or when your numerical data is not normally distributed, or if you suspect a non-linear but consistent (monotonic) relationship, or when outliers are present.
* Example: The relationship between students' rank in class based on exam scores and their rank in popularity, or the relationship between a person's income bracket (ordinal) and their happiness score (ordinal).

3. Kendall's Tau Correlation Coefficient (Kendall's τ)

* What it Measures: Kendall's Tau is another non-parametric measure of the strength and direction of a monotonic relationship between two variables. Like Spearman, it works with ranks, but its calculation is based on the number of "concordant" and "discordant" pairs in the data.

* Concordant Pair: If (x1, y1) and (x2, y2) are two observations, they are concordant if (x1 < x2 and y1 < y2) or (x1 > x2 and y1 > y2).
* Discordant Pair: They are discordant if (x1 < x2 and y1 > y2) or (x1 > x2 and y1 < y2).

Key Characteristics/Assumptions:

* Monotonicity: Similar to Spearman, it assesses monotonic relationships.
* Non-Parametric: It's also a non-parametric test, robust to non-normal distributions and outliers.
* Interpretation: Kendall's Tau is often considered more robust and provides a more interpretable probability measure of concordance. Its values tend to be smaller than Spearman's for the same data.
* When to Use It: When you have ordinal data, especially when dealing with smaller sample sizes or data with many tied ranks. It's often preferred over Spearman when dealing with a large number of ties in the ranks.
* Example: Ranking of movie preferences by two different critics, or the agreement between two different methods of ranking job candidates.

In summary, Pearson measures linear relationships for continuous, normally distributed data. Spearman and Kendall both measure monotonic (rank-based) relationships and are more suitable for ordinal data, non-normally distributed data, or data with outliers. Kendall is often favored when dealing with tied ranks or smaller samples.
