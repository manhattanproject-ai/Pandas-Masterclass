## 4. Ordered Numerical Data

image.png

These functions in Pandas are used to work with ordered numerical data, meaning data that has a natural ranking or sorting order. Let's explain each with an example:

Assume df is:

image.png

1. df.nlargest(n=5, columns='column_name'):

Retrieves the rows with the largest values in the specified column.

Example: df.nlargest(n=2, columns='Salary') would return:

image.png

2. df.nsmallest(n=5, columns='column_name'):

Retrieves the rows with the smallest values in the specified column.

Example: df.nsmallest(n=2, columns='Salary') would return:

image.png

3. df['column_name'].rank():

Assigns a rank to each value in the specified column based on their ascending order.

Example: df['Age'].rank() would return:

image.png

This means Eve has the lowest age (rank 1), Alice is second youngest (rank 2), and so on.

Summary :

These functions are useful for finding extreme values, ranking data, and understanding the order of values within a dataset.

nlargest and nsmallest return the actual rows, while rank returns a series of ranks.

rank can handle ties (e.g., if two people have the same age) using various tie-breaking methods.

## 5. Binning / Discretize and Clipping Data

1. Discretize Data
2. Clip Data

image.png

These functions in Pandas are used to perform operations on numerical data, such as binning, quantile-based discretization, and clipping. Let's explain each with an example:

Assume df is:

image.png

1. pd.cut(df['column_name'], bins=4):

Divides the data in the specified column into bins of equal width.

bins=4 creates 4 equal-width bins.

Example: pd.cut(df['Salary'], bins=4) might return:

image.png

Categories (4, interval[float64, right]): [(41250.0, 48750.0] < (48750.0, 56250.0] < (56250.0, 63750.0] < (63750.0, 71250.0]]


This shows the salary ranges for each person based on the 4 bins.

2. pd.qcut(df['column_name'], q=4):

Divides the data into quantiles, ensuring roughly equal number of values in each bin.
q=4 creates 4 quantiles (quartiles in this case).
Example: pd.qcut(df['Salary'], q=4) might return:

image.png

Categories (4, interval[float64, right]): [(44999.999, 49999.999] < (49999.999, 55000.0] < (55000.0, 60000.0] < (60000.0, 70000.0]]


This shows the salary quartiles for each person.

3. df['column_name'].clip(upper=20000):

Limits the values in the column to a specified upper bound.
upper=20000 sets the upper limit to 20000.
Example: df['Age'].clip(upper=30) would return:

image.png

This caps the age of Dave at 30.

Summary:

pd.cut is useful for creating bins with equal width.

pd.qcut is useful for creating bins with roughly equal number of data points.

clip is useful for setting upper or lower bounds on data values.

## 7. Cummulative Operation

1. Cummulative sum
2. Cummulative product
3. Cummulative max
4. Cummulative min
5. Sum
6. Product
7. Max
8. Min

image.png


These functions in Pandas are used to perform cumulative operations on numerical data within DataFrames. Let's explain each with an example:

Assume df is:

image.png

1. df['column_name'].cumsum():

Calculates the cumulative sum of values in the specified column.
Example: df['A'].cumsum() would return:

image.png

Each value is the sum of itself and all preceding values in the column.

2. df['column_name'].cumprod():

Calculates the cumulative product of values in the specified column.
Example: df['B'].cumprod() would return:

image.png

Each value is the product of itself and all preceding values in the column.

3. df.groupby(['column_name'])['column_name'].cummax():

Calculates the cumulative maximum value within each group defined by the specified column.
Example: df.groupby(['B'])['A'].cummax() would return:

image.png

For each group of rows with the same value in column B, it calculates the running maximum of values in column A.

4. df.groupby(['column_name'])['column_name'].cummin():

Calculates the cumulative minimum value within each group defined by the specified column.
Example: df.groupby(['B'])['A'].cummin() would return:

image.png

For each group of rows with the same value in column B, it calculates the running minimum of values in column A.

Summary:

cumsum and cumprod are useful for tracking running totals and products.
cummax and cummin are useful for finding running maximums and minimums, especially within groups of data.

```python
# Create dummy dataset
data = {'Name': ['Rahul', 'Sanjay', 'Ravi', 'Anil', 'Dhruv' , 'Akash'], 'Math': [80, 75, 70, 77, 82 , 93],
        'Physics': [83, 78, 76, 79, 85 , 95], 'Chemistry': [80, 79, 71, 74, 89 , 96],'History': [84, 71, 70, 73, 81 , 90]}
df = pd.DataFrame(data , index=['Rahul', 'Sanjay', 'Ravi', 'Anil', 'Dhruv' , 'Akash'] , columns=('Math', 'Physics' , 'Chemistry' , 'History'))
df
```

```python
# Define custom function to calculate range
def get_range(row):
    # Find difference between maximum and minimum value
    range_val = row.max() - row.min()
    return range_val

# Apply custom function at columns axis (i.e., set axis=1)
df['Range'] = df.apply(get_range, axis=1)

df
```

```python
# Create dummy dataset
data = {'Name': ['Rahul', 'Sanjay', 'Ravi', 'Anil', 'Dhruv' , 'Akash'], 'Math': [80, 75, 70, 77, 82 , 93],
        'Physics': [83, 78, 76, 79, 85 , 95], 'Chemistry': [80, 79, 71, 74, 89 , 96],'History': [84, 71, 70, 73, 81 , 90]}
df = pd.DataFrame(data , index=['Rahul', 'Sanjay', 'Ravi', 'Anil', 'Dhruv' , 'Akash'] , columns=('Math', 'Physics' , 'Chemistry' , 'History'))
df
```

```python
# Define custom function to calculate range
# result_type as broadcast replaces the actual value with the output
def get_range(row):
    range_val = row.max() - row.min()
    return range_val

# Broadcast results to all columns
df_output = df.apply(get_range, axis=1, result_type='broadcast')

df_output
```

```python
# Define custom function to store min and max value
# result_type as expand only has the new columns
def get_range_list(row):
    output_list = [row.min(), row.max()]
    return output_list

# Broadcast results to all columns
df_output = df.apply(get_range_list, axis=1, result_type='expand')

# Rename columns for better visibility
df_output.columns = ['Min', 'Max']

df_output
```

```python
# Define custom function to store min and max value
# result_type as reduce gives result opposite to expand
def get_range_list(row):
    output_list = [row.min(), row.max()]
    return output_list

# Broadcast results to all columns
df_output = df.apply(get_range_list, axis=1, result_type='reduce')

df_output
```

```python
# Define custom function to get difference from total (100)
def get_diff_from_100(x):
    return 100 - x

# Perform element-wise operation with applymap()
df_output = df.applymap(get_diff_from_100)

print(df_output)
```
