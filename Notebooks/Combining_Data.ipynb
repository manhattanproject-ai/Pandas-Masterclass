# 2. Concatenation

Imagine the table in the image is a collection of data, like a spreadsheet.

1. Rows: Represent individual records or entries (like information about different people).
2. Columns: Represent different attributes or features (like ID, Gender, DOB, City, Postal).
What is Concatenation?

Concatenation in Pandas is like "gluing" or "stacking" two or more DataFrames (which are essentially tables) together. You can do this in two main ways:

Concatenating Along Rows (Stacking Vertically):

1. This is like adding more rows to the bottom of your existing table.
2. You're essentially appending new data records.
3. In Pandas, this is done along axis=0.

Concatenating Along Columns (Stacking Horizontally):

1. This is like adding more columns to the right of your existing table.
2. You're essentially adding more attributes or features.
3. In Pandas, this is done along axis=1.

Key Points

1. axis Parameter: The axis parameter determines whether to concatenate along rows (0) or columns (1).
2. Index Alignment: Pandas will align the DataFrames based on their indexes (row labels) or column names, depending on the axis.
3. Missing Values: If the DataFrames have different columns or indexes, Pandas may introduce missing values (NaN) to fill in the gaps.

Summary

Concatenation is like combining multiple spreadsheets into one. You can either stack them on top of each other (adding more rows) or place them side by side (adding more columns). Pandas provides the pd.concat() function to do this easily.

# 3. Merging data

Imagine you have two tables (DataFrames) with information about customers and their orders.

Customers Table: Contains customer IDs and their details.
Orders Table: Contains order IDs and the customer IDs who placed them.
Merging in Pandas: Combining Tables

Merging is like combining information from two different tables based on a common column (or set of columns). The type of join determines how the rows are combined.

Types of Joins

1. INNER JOIN:

What it does: Returns only the rows where the matching values exist in both tables.

Analogy: It's like finding the customers who have placed orders.

Example:

If you merge the Customers and Orders tables using an INNER JOIN on CustomerID, you'll only get rows for customers who have placed orders.

Image Relevance: If the image had two tables with a common ID column but different sets of rows, an INNER JOIN would only keep the rows with matching ID values in both tables.

2. LEFT (OUTER) JOIN:

What it does: Returns all the rows from the left table and the matching rows from the right table. If there's no match in the right table, it fills the missing values with NaN.

Analogy: It's like finding all the customers and their orders, even if some customers haven't placed any orders.

Example:

If you merge the Customers and Orders tables using a LEFT JOIN on CustomerID, you'll get all customers, and their orders (if any). If a customer has no orders, the order columns will have NaN.

Image Relevance: If the left table in the image had rows with IDs that weren't in the right table, a LEFT JOIN would include those rows with NaN in the X3 and X4 columns.

3. RIGHT (OUTER) JOIN:

What it does: Returns all the rows from the right table and the matching rows from the left table. If there's no match in the left table, it fills the missing values with NaN.

Analogy: It's like finding all the orders and the customers who placed them, even if some orders don't have a matching customer.

Example:

If you merge the Customers and Orders tables using a RIGHT JOIN on CustomerID, you'll get all orders, and the customers who placed them (if any). If an order has no matching customer, the customer columns will have NaN.

Image Relevance: Similar to LEFT JOIN, but the focus is on the right table.

4. FULL (OUTER) JOIN:

What it does: Returns all the rows from both tables. If there's no match in either table, it fills the missing values with NaN.

Analogy: It's like finding all the customers and all the orders, regardless of whether they have a match.

Example:

If you merge the Customers and Orders tables using a FULL JOIN on CustomerID, you'll get all customers and all orders. Missing values will be filled with NaN where there are no matches.
Image Relevance: If the two tables in the image had different sets of ID values, a FULL JOIN would include all rows from both tables, with NaN in the missing columns.

5. CROSS JOIN:

What it does: Returns the Cartesian product of the rows from both tables. This means it combines every row from the first table with every row from the second table.

Analogy: It's like pairing every customer with every possible order, regardless of whether they actually placed that order.

Example:

If you merge the Customers and Orders tables using a CROSS JOIN, you'll get a table with every possible combination of customers and orders.

Image Relevance: A CROSS JOIN would create a table with 16 rows (4 rows from the left table x 4 rows from the right table), with every possible combination of rows.

## a. Combine_first

Understanding combine_first

combine_first in Pandas is used to fill missing values (NaN or null) in one DataFrame with values from another DataFrame, based on matching indices. It prioritizes the values from the first DataFrame.

How it Works

Matching Indices: It aligns the DataFrames based on their index (row labels).

Filling Missing Values: For each element in the first DataFrame:

If the element is not NaN, it's kept as it is.

If the element is NaN, it's replaced with the corresponding element from the second DataFrame (if it's not NaN).

# b. Combine

Understanding combine

combine in Pandas is a powerful function that allows you to apply a custom function to combine elements from two DataFrames based on their indices. It gives you fine-grained control over how the values are combined.

How it Works

Matching Indices: It aligns the DataFrames based on their index (row labels).

Applying a Custom Function: It applies a user-defined function to corresponding elements from both DataFrames.

Returning a Result: The function should return a single value that will be placed in the resulting DataFrame.

image.png

Summary

Combine is like having a tool that lets you compare two lists (or columns) side by side and apply a custom rule (your function) to decide how to combine the corresponding items. It's very flexible and allows you to perform complex comparisons and transformations.

```python
# DataFrame df_store_a
data_a = {
    'fruit_name': ['Cavendish Banana 1kg', 'NZ Fuji Apple 900g', 'Spain Black Plum 500g', 'Salustiana Orange 800g', 'Kyoho Grapes 450g'],
    'unit_price': [3.95, 5.95, 4.9, 5.2, 14.95]
}
df_store_a = pd.DataFrame(data_a)

# DataFrame df_store_b
data_b = {
    'fruit_name': ['Cavendish Banana 1kg', 'NZ Fuji Apple 900g', 'Spain Black Plum 500g', 'Salustiana Orange 800g', 'Kyoho Grapes 450g'],
    'unit_price': [4.35, 5.95, 3.9, 5.9, 11.95]
}
df_store_b = pd.DataFrame(data_b)
```

```python
df_store_a
```

```python
df_store_b
```

```python
# Execute combine() to find higher price
df_combi = df_store_a.combine(other=df_store_b,
                              func=np.maximum)

print(df_combi)
```

```python
# Execute combine() to find lower price
df_combi = df_store_a.combine(other=df_store_b,
                              func=np.minimum)

print(df_combi)
```

```python
# Custom function to calculate the average price

def calculate_average_price(price_a, price_b):
    return (price_a + price_b) / 2

# Combine the DataFrames using the custom function
combined_df = df_store_a['unit_price'].combine(df_store_b['unit_price'], calculate_average_price)

combined_df
```

```python
# Custom function to calculate the difference in prices

def calculate_difference(price_a, price_b):
    return (price_a - price_b)

# Combine the DataFrames using the custom function
combined_df = df_store_a['unit_price'].combine(df_store_b['unit_price'], calculate_difference)

combined_df
```

```python
# Custom function to compare prices and return a message

def compare_prices(price_a, price_b):

    if price_a < price_b:
        return f"Price increased from {price_a} to {price_b}"
    elif price_a > price_b:
        return f"Price decreased from {price_a} to {price_b}"
    else:
        return f"Price remains the same: {price_a}"

# Combine the DataFrames using the custom function
combined_df = df_store_a['unit_price'].combine(df_store_b['unit_price'], compare_prices)
combined_df
```

Combine Vs Combine_first

Summary

combine(): Is like having a tool that lets you compare two lists (or columns) side by side and apply a custom rule (your function) to decide how to combine the corresponding items. It's very flexible and allows you to perform complex comparisons and transformations.

combine_first(): Is like having two "fill in the blanks" forms. You start with the first form. If there are any blank spaces (NaNs), you fill them with the corresponding information from the second form. If there's still a blank space in the second form, it remains blank.

# c. Update

Understanding update()

The update() function in Pandas is used to modify the values in one DataFrame with values from another DataFrame, based on matching indices. It's essentially an in-place update operation, meaning it directly changes the original DataFrame.

How it Works

Matching Indices: It aligns the DataFrames based on their index (row labels).
Overwriting Values: For each element in the first DataFrame:
If the corresponding element in the second DataFrame is not NaN, the value in the first DataFrame is replaced with the value from the second DataFrame.
If the corresponding element in the second DataFrame is NaN, the value in the first DataFrame is not changed.

```python
# Update unit prices of Store B
df_store_b.update(df_store_a)
```
