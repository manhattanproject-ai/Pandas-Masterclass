# 1. Basic String Methods -

1. |df.astype('string')| It adds the specified scalar value to every element within the DataFrame df.|
2. |df['column_name'].str.len()| It obtains the number of characters in the string values within columns.|
3. |df['column_name'].str.lower()| It converts all strings in the column_name of the DataFrame df to lowercase.|
4. |df['column_name'].str.upper()| It converts all strings in the column_name of the DataFrame df to uppercase.|
5. |df['column_name'].str.title()| It returns a title-cased version of the string where every word starts with an uppercase character.|
6. |df['column_name'].str.capitalize()| It returns only uppercases the first character of the entire string, regardless of the number of words present.|
7. |df['column_name'].str.swapcase()| It is used to convert every character in a string to a case that is opposite of its current one.|
8. |df['column_name'].str.casefold()| It is used to returns a case-folded copy of the string, which can be useful for caseless matching.|
9. |df['column_name'].str.isalnum()| Checks whether all characters are alphanumeric in the column.|
10. |df['column_name'].str.isalpha()| Checks whether all characters are alphabetic in the column.|
11. |df['column_name'].str.isdecimal()| Checks whether all characters are decimal in the column.|
12. |df['column_name'].str.isdigit()| Checks whether all characters are digits in the column.|
13. |df['column_name'].str.islower()| Checks whether all characters are lowercase in the column.|
14. |df['column_name'].str.isnumeric()| Checks whether all characters are numeric in the column.|
15. |df['column_name'].str.isspace()| Checks whether all characters are whitespace in the column.|
16. |df['column_name'].str.istitle()| Checks whether all characters are titlecase in the column.|
17. |df['column_name'].str.isupper()| Checks whether all characters are uppercase in the column.|



# 2. Basic understanding of Text Data Type

### A. Create sample dataset

```python
import pandas as pd

data = {
    "Product Name": ["Logitech Brio 4K Webcam", "EMEET 1080P Webcam with Microphone", "NexiGo N60 1080P Web Camera"],
    "Retail Price": [130, 39, 50]
}

df = pd.DataFrame(data)
df
```

```python
# Obtain string values from existing DataFrame

str_values = df['Product Name'].values
str_values
```

```python
# Method 1 - StringDtype pandas data type
print('Method 1 output')

method_1_series = pd.Series(str_values, dtype=pd.StringDtype())
print(method_1_series)

# Method 2 - string alias
print('=' * 35)
print('Method 2 output')

method_2_series = pd.Series(str_values, dtype='string')
print(method_2_series)

# Convert existing column to string dtype
df['Product Name Converted'] = df['Product Name'].astype('string')

# View output
print(df['Product Name Converted'].dtype)
df.head()
```

# 3. String Methods for Basic transformations and checks

### A. Create sample data

```python
data = {
    "customer_id": ["264-42-4576", "763-23-7634", "763-55-7634"],
    "title": ['Mr', 'Ms', 'Dr'],
    "first_name": ['Raj', 'Neha', 'Ajay'],
    "last_name": ['martin', 'james', 'jones'],
    "ip_address": ['82.48.155.48/13', '95.48.177.48/18', '82.48.134.48/5'],
    "email": ['rmartin@james.edu', 'njames@gmail.com', 'ajones@yahoo.com'],

}

df = pd.DataFrame(data)
df
```

### B. Changing the datatype

```python
print(df.dtypes)
df = df.astype('string')
print(df.dtypes)
```

### C. Basic transformations in string

```python
#1. Get length of values in "title" column
df['title_length'] = df['title'].str.len()

# Display output
df[['title', 'title_length']]
```

```python
# 2. lower and uppercase of first_name column

# Lower case first name
df['first_name_lower'] = df['first_name'].str.lower()

# Upper case last name
df['last_name_upper'] = df['last_name'].str.upper()

# Display output
df[['first_name', 'first_name_lower', 'last_name', 'last_name_upper']]
```

```python
# 3. Create full name column

df['full_name'] = df['first_name'] + ' ' + df['last_name']

# Lower case the full name
df['full_name'] = df['full_name'].str.lower()

# Apply titlecase
df['full_name_title'] = df['full_name'].str.title()

# Apply capitalization
df['full_name_capitalize'] = df['full_name'].str.capitalize()

# View output
df[['full_name', 'full_name_title', 'full_name_capitalize']]
```

```python
# 4. Swap case of first name

df['first_name_swap'] = df['first_name'].str.swapcase()

# Swap case of last name
df['last_name_swap'] = df['last_name'].str.swapcase()

# Display output
df[['first_name', 'first_name_swap', 'last_name', 'last_name_swap']]
```

```python
# 5. Create new column for demo string values to apply casefold
df['demo'] = pd.Series(['', '', ''])

# Apply casefold method
df['demo_casefolded'] = df['demo'].str.casefold()

df
```

```python
# 6. Applying string Check Methods

# Check isnumeric for customer_id
df['customer_id_isnumeric'] = df['customer_id'].str.isnumeric()

# Check istitle for title
df['title_istitle'] = df['title'].str.istitle()

# Check isalpha for first_name
df['first_name_isalpha'] = df['first_name'].str.isalpha()

# Check isalnum for last_name
df['last_name_isalnum'] = df['last_name'].str.isalnum()

# Check isdigit for ip_address
df['ip_address_isdigit'] = df['ip_address'].str.isdigit()

# Check islower for email
df['email_islower'] = df['email'].str.islower()

# View output
df[['customer_id', 'customer_id_isnumeric',
          'title', 'title_istitle',
          'first_name', 'first_name_isalpha',
          'last_name', 'last_name_isalnum',
          'ip_address', 'ip_address_isdigit',
          'email', 'email_islower'
        ]]
```
